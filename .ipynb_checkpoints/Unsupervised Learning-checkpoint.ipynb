```python
# Import Dependancies
import pandas as pds
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Read Data
df = pds.read_csv("crypto_data.csv")
```

```python
# Read In CSV & Clean Data
df = df.loc[df['IsTrading'] == True]
df = df.loc[df['TotalCoinsMined'] > 0]
df["TotalCoinSupply"] = df["TotalCoinSupply"].astype(float)
df = df.dropna()
df = df.reset_index()
df.drop(['index', 'Unnamed: 0', 'CoinName','IsTrading'],axis='columns', inplace=True)
df = df.drop_duplicates()
```

```python
df.head()
```

```python
# Make Dummies
df = pds.get_dummies(df)
df
```

```python
# Scale Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['TotalCoinsMined', 'TotalCoinSupply']])
df_scaled = pds.DataFrame(scaled_data, columns=df.columns[:2])
df_scaled
```

```python
# Put Columns into Scaled DataFrame
for x in df.columns[2:]:
    df_scaled[x] = df[x].tolist()
df_scaled
```

```python
# Initalize PCA
pca = PCA(n_components=.9)
crypto_pca = pca.fit_transform(scaled_data)
df_pca = pds.DataFrame(data=crypto_pca, columns=["principal component 1", "principal component 2"])
df_pca.head()
pca.explained_variance_ratio_
```

```python
# TSNE
tsne = TSNE(learning_rate=40)
tsne_features = tsne.fit_transform(df)
```

```python
# Cluster With k-Means
inertia = []
k = list(range(1, 11))
#calculate the inertia in range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_pca)
    inertia.append(km.inertia_)
#creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pds.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Predicting Clusters With k=4
# Initialize k-Means Model
model = KMeans(n_clusters=4, random_state=0)

# Fit The Model
model.fit(df_pca)

# Predicting The Clusters
clusters = model.predict(df_pca)

# Adding The Predicted Class Column to DF
df_pca["class"] = model.labels_
df_pca.head()
```

```python
# Visualize The Clusters
plt.scatter(x=df_pca['principal component 1'], y=df_pca['principal component 2'], c=df_pca['class'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Crypto Clusters')
plt.show()
```
